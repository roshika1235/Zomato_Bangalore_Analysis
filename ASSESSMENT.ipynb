## checking null values of given dataset

```python
df.isnull().sum()
```

## dropping unwanted columns

```python
columns={"url","address","phone","reviews_list","rest_type","location"}
df=df.drop(columns,axis=1)
df
```

## changing boolean datatype for better analysis

```python
df["online_order"].unique()
df["online_order"]=df["online_order"].str.replace("Yes","1")
df["online_order"]=df["online_order"].str.replace("No","0")
df["online_order"].unique()
df["online_order"]=df["online_order"].astype(int)
df["online_order"].unique()
```

```python
# book_table
df["book_table"].unique()
df["book_table"]=df["book_table"].str.replace("Yes","1")
df["book_table"]=df["book_table"].str.replace("No","0")
df["book_table"].unique()
df["book_table"]=df["book_table"].astype(int)
df["book_table"].unique()
```

## changing rating in proper value and converting into float datatype

```python
df["rate"].unique()
df["rate"]=df["rate"].replace("/5","",regex=True)
df["rate"]=df["rate"].replace("NEW","0")                #replacing NEW,-,nan into 0
df["rate"]=df["rate"].replace("-","0")
df["rate"]=df["rate"].astype(float)                 #changing datatype into float
mean=round(df["rate"].mean())
df["rate"]=df["rate"].fillna(mean)
df["rate"].isnull().sum()
df["rate"].unique()
```

```python
df.info()
```

## handling null values for all columns

```python
df.isnull().sum()
```

## applied mode because it is categorical data


```python
df["dish_liked"].unique()
df["dish_liked"].mode()[0]
df["dish_liked"]=df["dish_liked"].fillna(df["dish_liked"].mode()[0])
df["dish_liked"].isnull().sum()
```

```python
df["cuisines"].unique()
df["cuisines"].mode()[0]
df["cuisines"]=df["cuisines"].fillna(df["cuisines"].mode()[0])
df["cuisines"].isnull().sum()
```

## applied median to avoid outliers

```python
df['approx_cost(for two people)'].unique()
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(",","")
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace("nan","")
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(float)
m=df['approx_cost(for two people)'].median()
df['approx_cost(for two people)']=df['approx_cost(for two people)'].fillna(m)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(int)
df['approx_cost(for two people)'].unique()
```

## Task 2: Univariate Analysis

```python
#Top 5 locations with highest number of restaurants
sns.histplot(x=df["listed_in(city)"],data=df,color="red")
plt.xticks(rotation=90)
plt.xlabel("city names")
plt.ylabel("number of restaurants")
plt.title("Top 5 locations with highest number of restaurants")
plt.show()
```

## location of
- BTM,
- koramangala 7th block ,
- koramangala 5th block  ,
- koramangala 4th block ,
- koramangala 6th block
has highest number of restaurants

```python
# Distribution of restaurant ratings
sns.histplot(x=df["rate"],data=df,color="red" ,kde=True)
plt.xlabel("rating")
plt.ylabel("number of restaurants")
plt.title("Distribution of restaurant ratings")
```

## most of the restaurants have rating betweem 3.5 to 4 , but some restaurants have 0 has their rating .
## 4 is the highest rating.
## and no customers rated 1,2 and 5


```python
#Percentage of restaurants offering online orders
plt.pie(df["online_order"].value_counts(),labels=df["online_order"].unique(),autopct="%.2f%%" ,colors=["red","grey"])
plt.title("Percentage of restaurants offering online orders")
plt.legend()
plt.show()
```

## 1 represent online_orders
## 0 represent offline
## mostly people order online with percent of 58.87
## there is only less difference so people prefer both equally likely.

```python
# Distribution of average cost for two
# Step 1: Calculate average cost_per_two for each location
location_avg_cost = df.groupby('listed_in(city)')['approx_cost(for two people)'].mean()

# Step 2: Plot histogram of location-wise average cost
plt.figure(figsize=(8,5))
sns.histplot(location_avg_cost, kde=True,color="red")
plt.xlabel("Average Cost for Two (Location-wise)")
plt.ylabel("Number of Locations")
plt.title("Distribution of Location-wise Average Cost for Two")
plt.show()
```

## mostly in 10-12 locations  they have higher average cost
